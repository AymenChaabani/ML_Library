```python
import numpy as np
from simpleann import SIMPLEANN,SIMPLEANNMINIBATCH
import matplotlib 
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets as ds
```

```python
file = open("./wine.data", "r")
a=file.read()
b=a.split("\n")
len(b)
c = []
for B in b : c.append(B.split(","))
c.pop(-1)
```

```python
D = [[float(z) for z in l ]for l in c]
```

```python
D = np.array(D)
```

```python
y = D[:,0].astype(int)
X = D[:,1::].astype(float)
y -= 1
X
```

```python
np.random.seed(1)
X, y = ds.make_moons(200, noise=0.20)
plt.scatter(X[:,0], X[:,1], s=40, c=y, cmap=plt.cm.Spectral)
```

```python
num_examples=len(X)
num_examples
```

```python
X.shape
```

```python
a=SIMPLEANN(13,3,15,1,100)
```

```python
b=SIMPLEANNMINIBATCH(13,3,4,1,50,10000)
```

```python
a.learn(X,y,False,False)
```

```python
b.learn(X,y,False,False)
```

```python
a.model
```

```python
b.model
```

```python
y_predict = a.infer(X)
y_predict
```

```python
yb_predict = b.infer(X)
```

```python
l = [ 1 if y[i]==y_predict[i] else 0 for i in range(len(y))]
```

```python
lb = [ 1 if y[i]==yb_predict[i] else 0 for i in range(len(y))]
```

```python
accuracy = np.sum(l)/num_examples*100
accuracy
```

```python
accuracyb = np.sum(lb)/num_examples*100
accuracyb
```
