```python
labels_test=np.zeros([25,1])
```

```python
j=0
k=0
for i in range(100):
    if i in idenxs:
        train[j]=cc[0][i]
        labels_train[j]=cc[1][i]
        j+=1
    else :
        test[k]=cc[0][i]
        labels_test[k]=cc[1][i]
        k+=1
        
        
```
